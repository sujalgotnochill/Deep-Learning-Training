```python
data.Gender.replace({'Male':1,'Female':0},inplace=True)
```

```python
data.OverTime.replace({'Yes':1,'No':0},inplace=True)
```

```python
data.Attrition.replace({'Yes':1,'No':0},inplace=True)
```

```python
data.BusinessTravel.replace({'Non-Travel':0,'Travel_Rarely':1,'Travel_Frequently':2},inplace=True)
```

```python
data.Department.replace({'Cardiology':0,'Maternity':1,'Neurology':2},inplace=True)
```

```python
data.EducationField.replace({'Life Sciences':0,'Technical Degree':1,'Medical':2,'Marketing':3,'Human Resources':4,'Other':5},inplace=True)
```

```python
data.MaritalStatus.replace({'Single':0,'Married':1,'Divorced':2},inplace=True)
```

```python
data.JobRole.replace({'Admin':0,'Nurse':1,'Therapist':2,'Administrative':3,'Other':4},inplace=True)
```

```python
scale_list = ["Age","DailyRate","DistanceFromHome","HourlyRate","MonthlyIncome","MonthlyRate","PercentSalaryHike","TotalWorkingYears","YearsAtCompany","YearsInCurrentRole","YearsSinceLastPromotion","YearsWithCurrManager"]
```

```python
def minmax(d,l):
  for cols in l:
    d[cols] /= max(d[cols])
```

```python
X = data.drop('Attrition',axis=1)
```

```python
X
```

```python
y = data.Attrition
```

```python
y
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size= 0.2)
```

```python
model=keras.Sequential([ keras.layers.Dense(100, activation="softmax"),
                         keras.layers.Dense(50, activation='softmax'),
                         keras.layers.Dense(2, activation='sigmoid')
                        ])
```

```python
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"]
              )
```

```python
model.fit(X_train , y_train , epochs=100)
```

```python
model.evaluate(X_test , y_test)
```
